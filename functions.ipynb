```python
def introduce(name,address,occupation):
    """This function introduces a person and their details."""
    print(f"My name is {name}.I live in {address}and i work as {occupation}")
```

```python

#calling function with parameters multile using positional arguments
introduce("Sita","Kathmandu","Engineer")
introduce("man","doctor","hello")
```

```python
#eg of args
def sum_all(*args):
    """This function returns sum of all arg"""
    total= sum(args)
    print(f"Sum: {total}")
```

```python
sum_all(1,2,3,4)
```

```python
def display_info(**kwargs):
    """This function display information passed as keyword arguments"""
    for key, value in kwargs.items():
        print(f"{key}: {value}")
```

```python
print("\n Displaying using **kwargs")
display_info(name="Ram",age="40")
```

```python
#Returns statement sends values back to the caller

def calculate(a,b):
    """This fun returns the sum , diiference,products division of to numbers"""
    sum_result = a+b
    diff_value = a-b
    prod_value =a*b
    division_result =a/b
    return sum_result, diff_value, prod_value, division_result
```

```python
fno = int(input("Enter number"))
sno = int(input("Enter 2nd number"))
sum_val,diff_val,prod_val,div_val = calculate(fno,sno)
print(f"Sum:{sum_val},Difference:{diff_val},product:{prod_val},Division:{div_val}")
```

```python
def grading():
    marks = list(map(int, input("Enter marks of 5 subjects separated by space: ").split()))
    total = sum(marks)
    percentage = total / 5
    print(f"Total: {total}")
    print(f"Percentage: {percentage}%")

    if percentage >= 90:
        print("Excellent!")
    elif percentage >= 75:
        print("Good job!")
    elif percentage >= 50:
        print("You passed.")
    else:
        print("Try harder next time.")

grading()
```
