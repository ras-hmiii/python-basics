```python
import numpy as np
import math
import time

data_size = 1000000

python_data = list(range(1,data_size+1))
start_time = time.time()
python_sqrt = [math.sqrt(x) for x in python_data]
end_time = time.time()
print(f"Python list sqrt time: {end_time - start_time:.6f} seconds")

#Using Numpy 
numpy_data = np.arange(1,data_size + 1)
start_np=time.time()
numpy_sqrt = np.sqrt(numpy_data)
end_np=time.time()
print(f"Numpy array sqrt time: {end_np - start_np:.6f} seconds")

```
