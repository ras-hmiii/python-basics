```python
#Random Numbers Generation
import numpy as np

np.random.seed(0) # For reproducibility
uniform = np.random.rand(5)
normal =np.random.randn(5)
intergers =np.random.randint(1,200,size=5)
print(uniform)
print(normal)
print(intergers)
```

```python
noraml_dist  = np.random.normal(loc=20, scale=5, size=100)
print(noraml_dist)
```

```python
print(f"Mean: {np.mean(noraml_dist)}, Std Dev: {np.std(noraml_dist)}")
```
