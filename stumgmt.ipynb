```python
class StudentMgmt:
    def __init__(self, student_name,student_id):
        self.student_name = student_name,
        self.student_id = student_id

    def display_student_info(self):
        print(f"Student Name: {self.student_name}")
        print(f"Student ID: {self.student_id}")
    def create_student_record(self):
        print(f"Creating record for {self.student_name} with ID {self.student_id}")
    def update_student_record(self, new_name=None, new_id=None):
        if new_name:
            self.student_name = new_name
        if new_id:
            self.student_id = new_id
        print(f"Updated record to Name: {self.student_name}, ID: {self.student_id}")

    def delete_student_record(self):
        print(f"Deleting record for {self.student_name} with ID {self.student_id}")
        self.student_name = None
        self.student_id = None
        
```
